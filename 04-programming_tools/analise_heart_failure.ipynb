Criar um dicionário para construir o novo dataframe _visao_geral_ é mais seguro, pois, quando o kernell reiniciar e a informação não estiver no clipboard, a célula acima não irá ler os dados corretos e o seu dataset ficará comprometido.

Um bom jeito de criar o dicionário a partir destas informações (colunas, valores não nulos e tipo de variável) é o seguinte:

```python
# Identificar as colunas com o  parâmetro columns
columns = df.columns
# Identificar os valores não nulos pela soma dos valores encontrados com o método notnull:
non_null = df.notnull().sum()
# Identificar os tipos de variáveis com o parâmetro dtypes
types = df.dtypes

# Transformar todos em lista usando o método tolist():
columns = columns.tolist()
non_null = non_null.tolist()
types = types.tolist()

# Criar um dicionário com estas variáveis e transformá-lo em um dataframe:
dicionario = {'Columns': columns, 'Non-Null Count': non_null, 'Dtype': types}
visao_geral = pd.DataFrame(dicionario)


display(visao_geral)
```

#### 1.1.2. _Tipos de variáveis e seu preenchimento_

Na estatística, as variáveis podem ser classificadas em **quantitativas** ou **qualitativas (categóricas)**.

As **variáveis quantitativas** são aquelas que podem ser medidas em uma escala quantitativa, ou seja, por valores numéricos que façam sentido, podendo ser subclassificadas em discretas ou contínuas, a depender de se faz sentido que seus valores sejam fracionados ou não.

Já as **variáveis qualitativas ou categóricas** são características categorizáveis, não definidas por valores numéricos. Podem ser nominais ou ordinais, quando progressivas.

Retornando às informações do dataset, é possível verificar a presença de algumas variáveis categóricas como sexo (_sex_) e se o paciente é fumante (_smoking_). Apesar disso, vê-se que o seu preenchimento foi feito com números inteiros (_int64_):

### .sort_values() - Method

```python
titanic
```

```python
titanic.sort_values("Name")
```

#### _Ascending or Descending_

Setting the **ascending argument** to False will sort the data the other way around. 

```python
titanic.sort_values("Name", ascending=False)
```

#### _Sort by Multiple Variables_

We can **sort by multiple variables** by passing a list of column names to sort_values.

```python
titanic.sort_values(["Age", "Name"])
```

You can adjust ascending or descending arguments to each column:

```python
titanic.sort_values(["Age", "Name"], ascending=[False, True])
```

### df["column"]

It's a way to show only one or some columns.

```python
titanic["Name"]
```

#### _Selecting multiple columns_

You have to put the column names between brackets, as an array or a list.

```python
titanic[["Name", "Age"]]
```

```python
cols_to_subset = ["Name", "Age"]
titanic[cols_to_subset]
```

#### _Filter rows_

You can set a condition to show only some of the rows that meets this condition.

```python
filter = titanic["Age"] > 50
print(filter)
```

```python
titanic[filter]
```

Same as:

```python
titanic[titanic["Age"] > 50]
```

You can do it with strings:

```python
filter2 = titanic["Sex"] == "female"
```

```python
titanic[filter2]
```

Same as:

```python
titanic[titanic["Sex"] == "female"]
```

### _Logical Operators & |_

You can subset based on multiple conditions using operators & (AND) and | (OR)

```python
titanic[filter & filter2]
```

```python
titanic[filter | filter2]
```

#### .isin() - Method

 This takes in a list of values to filter for. 

```python
selected_cabins = ['C85', 'B42']
cabin_filter = titanic["Cabin"].isin(selected_cabins)
titanic[cabin_filter]
```
